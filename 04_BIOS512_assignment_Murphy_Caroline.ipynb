Use `read_csv` to load the data from the TidyTuesday GitHub repository. The project URL is can be found at [https://github.com/rfordatascience/tidytuesday/blob/master/data/2020/2020-10-27/](https://github.com/rfordatascience/tidytuesday/blob/master/data/2020/2020-10-27/).

```r
library('tidyverse')
```

```r
wind_turbine=read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv') 

wind_turbine |> head() 
```

Each row in the table represents a single turbine.

Make a bar chart 📊 **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
wind_turbine_sorted=wind_turbine |> 
    mutate(province_territory_sorted=province_territory |> 
          fct_infreq() |> 
          fct_rev()) 

wind_turbine_sorted |> 
    head() 
```

```r
options(repr.plot.width = 5, repr.plot.height = 3, repr.plot.res = 300)

p=ggplot(wind_turbine_sorted, aes(x=province_territory_sorted)) + 
    geom_bar() + 
    coord_flip() + 
    labs(x="Province Territory", y="Turbines (N)")

p
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
wind_turbine_sorted_2=wind_turbine |> 
    mutate(manufacturer_sorted=manufacturer |> 
          fct_infreq() |> 
          fct_rev()) 

wind_turbine_sorted_2 |> 
    head() 
```

```r
as.data.frame(table(wind_turbine_sorted_2$manufacturer_sorted)/
                  sum(table(wind_turbine_sorted_2$manufacturer_sorted))) 
```

```r
p2=ggplot(wind_turbine_sorted_2, aes(x=manufacturer_sorted)) + 
    geom_bar() + 
    coord_flip() + 
    labs(x="Manufacturer", y="Turbines (N)")
p2
```

```r
wind_turbine_lumped=wind_turbine_sorted_2 |> 
    mutate(manufacturer_lumped=manufacturer |> 
          fct_lump(n=6) |> 
          fct_infreq() |> 
          fct_relevel('Other', after=Inf) |> 
          fct_rev() 
    ) 

p3=ggplot(wind_turbine_lumped, aes(y=manufacturer_lumped)) + 
    geom_bar() + 
    labs(x="Turbines (N)", y="Manufacturers")

p3
```
